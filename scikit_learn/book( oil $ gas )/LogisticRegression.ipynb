## [2] let's see the collerations between the features

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True,cmap="viridis",linewidth=1,linecolor="#fff")
plt.show()
```

## Degree of correlation:

- <span style="color:red">Perfect</span>: If the value is near Â± 1, then it said to be a perfect correlation: as one variable increases, the other variable tends to also increase (if positive) or decrease (if negative).
- <span style="color:red">High degree </span>: If the coefficient value lies between Â± 0.50 and Â± 1, then it is said to be a strong correlation.
- <span style="color:red">Moderate degree </span>: If the value lies between Â± 0.30 and Â± 0.49, then it is said to be a medium correlation.
- <span style="color:red">Low degree </span>: When the value lies below + .29, then it is said to be a small correlation.
- <span style="color:red">No correlation </span>: When the value is zero.

### so form the heatmap:
- there is no collinear features
- we will take all the features 

```python
X = df.drop("Quitting",axis=1)
Y=df["Quitting"]
Y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
lr=LogisticRegression(penalty='l2',C=1,solver='lbfgs')
# which is the defult parameters
```

```python
x_train,x_test,y_train,Y_test=train_test_split(X,Y,test_size=.3)
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

## model evaluation

```python
plt.figure()
sns.heatmap(confusion_matrix(Y_test,y_pred),annot=True,cmap="viridis")
# the horizontal is the predicted
# the virtical is the actual
```

```python
print(classification_report(Y_test,y_pred))
```

## apply the girdsearch

```python
from sklearn.model_selection import GridSearchCV
```

```python
params={"penalty":['l1', 'l2', 'elasticnet', 'none'],
         "C":np.linspace(.01,10,20),
        "solver":['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
       }
log=LogisticRegression()
log_cv=GridSearchCV(log,params,cv=5)
```

```python
log_cv.fit(x_train,y_train)
```

```python
log_cv.best_params_
```
